```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [5, 4, 3, 2, 1]
y2 = [1, 2, 3, 4, 5]

# Use a color map for getting colors
cmap = cm.get_cmap('tab10')

# Plot the lines with colors from the color map
plt.plot(x, y1, color=cmap(0), label='Line 1')
plt.plot(x, y2, color=cmap(1), label='Line 2')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```
